```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
print('imported--')
```

```python
df=pd.read_csv('finaldata.csv')
print('file reading done--')
```

```python
df.head(10)
```

```python
pd.options.display.max_columns=None
pd.options.display.max_rows=None


print(df.describe())
```

```python
for col in df.columns:
            indexNames = df[df[col] == 'inf'].index
            if not indexNames.empty:
                print('deleting {} rows with Infinity in column {}'.format(len(indexNames), col))
                df.drop(indexNames, inplace=True)
for col in df.columns:
            indexNames = df[df[col] == 'NaN'].index
            if not indexNames.empty:
                print('deleting {} rows with Infinity in column {}'.format(len(indexNames), col))
                df.drop(indexNames, inplace=True)
```

```python
col=['FlowBytess','FlowPacketss']
df.drop(col,inplace=True,axis=1)
print(df.describe())
```

```python
features=df.columns[:-2]
print(features)
print('number of features:{}'.format(len(features)))
```

```python
for feature in features:
    print(feature)
    order=df['Label.1'].unique()
    fig=plt.figure()
    fig.set_size_inches(15, 6)
    plt.rcParams.update({'font.size':15})
    sns.violinplot(x='Label.1',y=feature,data=df,order=order,cut=0)
    plt.title('Distribution Plot of \'{}\''.format(feature))
    plt.show()
    
```
