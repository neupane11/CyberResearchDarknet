```python
# remove space from column name
df.columns=df.columns.str.replace(' ', '')

#remove / from columnname
df.columns=df.columns.str.replace('[/]','')
```

```python
#identify columns that contains single value
print(df.nunique())
```

```python
# we can see many columns have single value
# Columns that have a single value for all rows do not contain any information for modeling.So drop it
# a single value means that each row for that column has the same value
```

```python
def DropSingleValueColumns():
    # get number of unique values for each column
    print(df.shape)
    for col in df.columns:  # Loop through columns
        if len(df[col].unique()) == 1:  # Find unique values in column along with their length and if len is == 1 then it contains same values
            df.drop([col], axis=1, inplace=True)  # Drop the column
    print(df.shape)
```

```python
DropSingleValueColumns()
```

```python
print(df.nunique())
```

```python
#see if rows has duplicate data
dups = df.duplicated()
# report if there are any duplicates
#print(dups.any())
# list all duplicate rows
#print(df[dups])
print(df.shape)
# delete duplicate rows
df.drop_duplicates(inplace=True)
print(df.shape)
```

```python
# drop rows if any NaN values
print(df.shape)
df.dropna()
print(df.shape)
```

```python
#convert string into numeric value
s=pd.Series(["FlowID","SrcIP","DstIP","Timestamp"])
df[["FlowID","SrcIP","DstIP","Timestamp"]]=df[["FlowID","SrcIP","DstIP","Timestamp"]].apply(pd.to_numeric,errors='coerce')
pd.to_numeric(s,errors='coerce')
```
